```python
import pandas as pd
sites = pd.read_csv("sites.csv")
```

```python
# Don't count gone sites.
sites = sites[sites['is_gone'] == False]
# "Public sites that have 0 courses are not counted; private instances are counted towards total sites"
sites = sites[(sites['is_private_instance'] == True) | (sites['course_count'] > 0)]
# Assume English.
sites.fillna(value={'languages': 'English'}, inplace=True)
sites = sites.drop(columns=['name', 'is_gone', 'updated', 'is_private_instance'])
sites['languages'] = sites['languages'].str.replace(' ', '')
sites['languages'] = sites['languages'].str.replace('English,', '')
sites['languages'] = sites['languages'].str.replace('French,', '')
```

```python
sites.groupby('languages').sum().sort_values('course_count', ascending=False)
```
